```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()

nltk.download('stopwords')

stop_words=set(stopwords.words("english"))
print(type(stop_words), len(stop_words))

import re
re.sub("ab",'*','absabdosba')

def rem_symbols(line):
  #return re.sub("[^A-Aa-z0-9\s]+",'', line)
  return re.sub('[^A-Za-z0-9\s]+', '', line).lower()  # do again

import csv
import sys
csv.field_size_limit(sys.maxsize)          # use it to read whole line no limit on line size

BLOCKSIZE= 10000
####################################################################################
import pandas as pd
A = pd.DataFrame({},columns=['Title','Author','info_abt_paper','text_block'])

def load_row(my_row):
  A.loc[len(A.index)+1] = my_row

with open("/content/drive/MyDrive/CRAN/Unzipped_cran/cran.all.1400", 'r') as x:
  #next(x) # it is used when first line contains column names
  full_string = x.read()
  full_string = full_string.replace('\n',' ').split(".I")[1:]
  for seg in full_string:
    seg = seg.replace(".T","_").replace(".A","_").replace(".B","_").replace(".W","_")
    seg=seg.split("_")[1:]
    load_row(seg[:4])
A.index.name = 'ID'
A=A.reset_index()

######################################################################################
```

```python

def bsbi(l):
  #freq_dist=  defaultdict(set)
  freq_dist = []
  current_block =0
  total_files = 0
  line_no =0
  #########################################
  for i,line in A.iterrows():
      line_no += 1
      ID,	Title,	Author,	info_abt_paper,	text_block = line
      if l=='Text':
        p= text_block
      elif l=='Title':
        p= Title
      elif l=='Info':
        p = info_abt_paper
      elif l =='Author':
        p= Author
      for word in p.split():                 # Splitting text_bloc to terms
        word = rem_symbols(word)
        if word and word not in stop_words:
          word = porter.stem(word)
          # so add  this 
          #t= freq_dist[word]
          #freq_dist[word]= t+[word]
          current_block += 1
          freq_dist.append((word,ID))
          """
          if word not in freq_dist:
            # so first time
            current_block += 1
          #if id not in freq_dist[word]:
          if not freq_dist[word].__contains__(ID):
            # the id is not added
            freq_dist[word].add(ID)
            current_block += 1"""
          # if word in freq_dist:
          #   # present
          #   t = freq_dist[word]
          #   freq_dist[word] = t+[ID]
          # else:
          #   # not present
          #   freq_dist[word] = [str(ID)]
        if current_block>= BLOCKSIZE:
          # IT IS FULL now  
          sorted_list = sorted(freq_dist, key= lambda _:_[0])
          with open(f'/content/drive/MyDrive/CRAN/Dumps_1/{l}_dump/OP{total_files}.txt', 'w') as  f:
            for word, doc_ids in sorted_list:
              f.write(word)
              for doc_id in doc_ids:
                f.write(f' {doc_id}')
              f.write('\n')  # writing one line is completed
            current_block = 0
            freq_dist.clear()
            total_files += 1
            #print(line_no, 'row done')  # so n lines copied to the new file
            # test
            #if total_files >= 5:
            #  return
            print("OP{}".format(total_files-1),'dumped','|',line_no,'rows done')
      # last time we have something in freq_dist but we have not dumped it 
      sorted_list = sorted(freq_dist, key = lambda _:_[0])
      with open(f'/content/drive/MyDrive/CRAN/Dumps_1/{l}_dump/OP{total_files}.txt','w') as f:
        # this is for last values
        # TODO:: DO IT BY FN SO NO REPEATATION OF CODE
        for word, doc_ids in sorted_list:
          f.write(word)
          #for doc_id in doc_ids:
          f.write(f' {doc_ids}')
          f.write('\n')  
        current_block = 0
        freq_dist.clear()
        total_files += 1
        #print(line_no, 'row done')
        print("OP{}".format(total_files-1),'dumped','|',line_no,'rows done')  
#c= ["Text","Title","Info","Author"] 
bsbi("Author") 
############################################################################################################
```

```python
def bsbi2(k):
  file_names = [f'/content/drive/MyDrive/CRAN/Dumps_1/{k}_dump/OP{i}.txt' for i in range(0,1400)]   # OP0.txt OP1.txt OP2.txt
  ################################################################################
  # Let's be manual
  my_dict={}
  from collections import defaultdict
  for file_path in file_names:
    with open(file_path,'r') as x:
      for i in x:
        a,b = i.split()
        #if a=='wave' and b=='1244':  # to check for wave text if it is capturing frequency
        #  print(a,b)
        if a in my_dict:
          # present
          t = my_dict[a]
          my_dict[a] = t+[b]
        else:
          # not present
          my_dict[a] = list(b)
  ####################################################################################################################
  with open(f'/content/drive/MyDrive/CRAN/Dumps_1/{k}_combine/merged_{k}.txt', 'w') as x:
    for term,id_list in my_dict.items():
      my_set = set(id_list)
      freq_dist = []
      freq_dist = [(i,str(id_list.count(i))) for i in my_set]
      x.write(term+" ")
      t = sum([int(j) for i,j in freq_dist])
      x.write(str(t)+" ")
      my_str_list = [a+"_"+b for a,b in freq_dist]
      my_str = " ".join(my_str_list)
      x.write(my_str+"\n")
#c= ["Text","Title","Info","Author"] 
bsbi2("Author")
```
